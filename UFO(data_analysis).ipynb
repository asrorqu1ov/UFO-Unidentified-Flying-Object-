# Dataset haqida malumot

Ushbu scrubbed.csv datasetimiz **UFO(Unidentified Flying Object)** yani **Aniqlanmagan Uchuvchi obyektlar** haqida 1906 yildan 2014 yilgacha bo'lgan hodisalar   yig'ilgan dataset.  

#### Datasetdagi ustunlar haqida malumot  
1. **datetime** - Hodisa sodir bo'lgan sana yani UFO hodisasi qachon qayerda ko'rganni bildiradi
2. **city** - UFO hodisasi sodir bo'lgan shahar
3. **state** - Shahar joylashgan shtat yokida viloyati 
4. **country** - Shahar joylashgan davlat 
5. **shape** - UFO shakli Guvohlar UFOni qanday shaklda ko'rgani
6. **duration (seconds)** - Hodisa qancha davom etgani (sekundlarda)
7. **duration (hours/min)** - Hodisa qancha davom etgani (soat va minutda)
8. **comments** - Guvohlar tomonidan yozilgan matnli izoh
9. **data posted** - Hodisa haqida malumot joylashtirilgan sana
10. **latitude** - Joyning shimol–janub koordinatasi
11. **longitude** - Joyning g‘arb–sharq koordinatasi  
  
#### Biz qo'shgan ustunlar 
1. **geolocation** - Joyning koordinatalarini birlashtirib bitta ustun qilib qo'ydik (latitude , longitudeni ustunlarni drop qilib yubordik )
2. **year** - datetime ustunidan yearni ajratib oldik
3. **month** - datetime ustunidan monthni ajratib oldik 
4. **day** -datetime ustunidan dayni ajratib oldik
5. **hour** -datetime ustunidan hourni ajratib oldik
6. **minute** - datetime ustunidan minuteni ajratib oldik
7. **season** - month ustuni bo'yicha fasllarga ajratib oldik bu tahlil uchun kerak va vizualizatsiyada ham 
8. **time_of_day** - hour ustuni boyicha hodisa qaysi payt sodir bo'lganni aniqladik night or day
9. **light_or_glow** - Bu hodisa sodir bo'lganda odamlar yozgan matn boyicha yorug'lik bo'lgan yoki bo'lmagani bilish uchun bu ustunni yaratdik vizualizatsiya ishlatildi  

#### O'chirilgan ustunlar 
1. **state** - Bundan ko'ra city va country ustunlaridan ko'proq malumotlar oldik bu keraksiz bo'lib qoldi va buni drop qildik 
2. **duration (hour/min)** - duration (seconds) ustuni borligi uchun bu ustun bizga leralsiz bo'lib qoldi va asosiy malumotni duration seconds ustunidan oldik 
3. **comments** - Hodisa sodir bo'lganda yorug'lik bo'lgan yoki yo'qligi bilganimizdan so'ng bu ustundan keraksiz bo'lib qoldi va drop qilib yubordik 
4. **latitude** - Koordinatalarni bitta ustunga birlashtirdik
5. **longitude** - Koordinatalarni bitta ustunga birlashtirdik

```python
df.columns = df.columns.str.strip().str.replace(' ', '_')
df.columns
```

```python
df = df.rename(columns={'duration_(seconds)': 'duration_second'})
```

```python
df['duration_second'] = pd.to_numeric(df['duration_second'], errors='coerce')
df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')
```

```python
df['shape'] = df['shape'].fillna('unknown').str.lower()
```

### Data transformation engineer

```python
df.info()
```

| Xususiyat | `apply()` | `transform()` |
| --- | --- | --- |
|  Maqsad | Har bir **qator** yoki **ustun**ga funksiya qo‘llash | Har bir **element**ga yoki **ustun bo‘yicha o‘zgartirish** |
|  Natija o‘lchami | Natija shakli o‘zgarishi mumkin (masalan, yangi ustunlar) | Natija har doim **kiruvchi o‘lcham bilan bir xil** |
|  Guruhlash (`groupby`) bilan ishlaganda | Har bir guruhga umumiy qiymat qaytaradi | Har bir **qator uchun** transformatsiya qilingan qiymat beradi |
|  Tipik qo‘llanish | `apply(lambda x: ...)` — maxsus hisob-kitoblar | `transform('mean')`, `transform(lambda x: (x - x.mean()))` — normalizatsiya yoki scaling uchun |

```python
df['duration_second'] = df.groupby('shape')['duration_second'].transform(lambda x: x.fillna(x.median()))
df['latitude'] = df.groupby('city')['latitude'].transform(lambda x: x.fillna(x.median()))
```

```python
df['latitude'].fillna(df['latitude'].median(), inplace=True)
```

```python
df['geolocation'] = list(zip(df['latitude'], df['longitude']))
```

zip - bir nechta iterablelarni birlashtirishda kerak bo'ladi va tuple juftligini qaytaradi 

```python
df.drop(['latitude', 'longitude'], axis=1, inplace=True)
df.head()
```

```python
df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')
#df['date_posted'] = pd.to_datetime(df['date_posted'], errors='coerce')
df.dtypes
```

```python
global_median = df['datetime'].median()

df['datetime'] = df.groupby('city')['datetime'].transform(lambda x: x.fillna(x.median() if not x.dropna().empty else global_median))
```

```python
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['hour'] = df['datetime'].dt.hour
df['minute'] = df['datetime'].dt.minute

df.head()
```

```python
df.dtypes
```

### Data Analyst

```python
def fasil(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Autumn'
    else:
        return None

df['season'] = df['month'].apply(fasil)
df.head()
```

```python
def kun_tun(hour):
    if hour >= 6 and hour < 18:
        return 'Day'
    else:
        return 'Night'

df['time_of_day'] = df['hour'].apply(kun_tun)
df.head()
```

```python
import re

def light_glow(comment):
    if isinstance(comment, str) and re.search(r'light|glow', comment, re.IGNORECASE):
        return 1
    else:
        return 0

df['light_or_glow'] = df['comments'].apply(light_glow)
df.head()
```

```python
df.drop(['duration_(hours/min)', 'comments', 'state'], axis=1, inplace=True)
df.columns
```

```python
df.head()
```

```python
df['country'].value_counts()
```

```python
# df['country'] = df.groupby('city')['country'].transform(lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else x))
df['country'].fillna(df['country'].mode()[0], inplace=True)
```

### Vizualizatsiya 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Vaqt bo'yicha analysis (lineplot, countplot)
> axes[0] - yillar davomida UFO hodisalar ko'paygan yoki kamaygani haqida malumot beradi  
> axes[1] - UFO hodisalari qaysi oylarda ko'p uchragani haqida malumot olamiz 

```python
fig , axes = plt.subplots(1, 2, figsize = (18,10))

# year boyicha countlarni bilish uchun qilayapmiz 
year_counts = df.groupby('year').size().reset_index(name='count')

sns.lineplot(data= year_counts, x='year', y='count',  ax=axes[0], color= 'blue' )
sns.countplot(data= df, x='month', ax= axes[1])


#title berish 
axes[0].set_title("UFO Sightings by Year")
axes[1].set_title("Monthly UFO Activity")


# umumiy title
fig.suptitle("UFO Sightings Temporal Patterns", fontsize=16, fontweight="bold")

plt.tight_layout()
plt.show()
```

#### Geografik tahlil (barplot)
* hodisalar qayerlarda ko'proq uchragani ko'rish uchun  
* eng ko'p hodisa bo'lgan top 10 shahar va davlatlar 

```python
top_10_city = df.groupby('city')['duration_second'].count().sort_values(ascending=False).head(10).reset_index(name='count')

top_5_country = df.groupby('country')['duration_second'].count().sort_values(ascending=False).head(5).reset_index(name='count')
```

```python
fig, axes = plt.subplots(1, 2, figsize= (18,10))

sns.barplot(data=top_10_city, x='city', y='count', palette='viridis', ax= axes[0])
sns.barplot(data=top_5_country, x='country', y='count', palette='magma', ax=axes[1])

#title berish 
axes[0].set_title("Top 10 Cities ", fontsize=13, fontweight='bold')
axes[1].set_title("Top 5 Countries", fontsize=13, fontweight='bold')

# oq nomlari
axes[0].set_xlabel("City")
axes[0].set_ylabel("Number of Sightings")
axes[1].set_xlabel("Country")
axes[1].set_ylabel("Number of Sightings")

# x oqidagi shahar nomlarini 45 gradusga burish 
axes[0].tick_params(axis='x', rotation=45)

#umumiy title
fig.suptitle("Geographical Distribution of UFO Sightings", fontsize=16, fontweight='bold', color='green')

plt.tight_layout()
plt.show()
```

#### Qanday shakldagi va davomiyligi boyicha (boxplot, barplot)
* Shapesdagi top 10 taligi
* Top shapesdagilarni duration vaqti boyicha boxplot 
* Top shapesdagi hodisa bo'lgandagi light_or_glow bo'lganligi

```python
fig, axes = plt.subplots(1, 3, figsize=(18,10))

sns.barplot(data= top_10_shape, x= 'shape', y='count', palette='viridis', ax=axes[0])
sns.boxplot(data=top_shapes, x='shape', y='duration_second', color = 'teal', ax=axes[1])
sns.boxplot(data= top_shapes, x='shape', y='duration_second',hue='light_or_glow', ax=axes[2])
#sns.violinplot(data=top_shapes, x='shape', y='duration_second', hue='light_or_glow', palette='Set2', ax=axes[2])

#title qoyish
axes[0].set_title("Top 10 Shapes ", fontsize=13, fontweight='bold')
axes[1].set_title("Duration of Top 10 UFO Shapes (log scale)", fontsize=13, fontweight='bold')
axes[2].set_title("Duration by Shape and Light/Glow (long scale)", fontsize=13, fontweight= 'bold')

# label qoyish
axes[0].set_xlabel("Shape")
axes[0].set_ylabel("Number of Sightings")
axes[1].set_xlabel("Shape")
axes[1].set_ylabel("Duration (seconds)")
axes[2].set_xlabel("Shape")
axes[2].set_ylabel("Duration (seconds)")


# x oqidagi shahar nomlarini 45 gradusga burish 
axes[0].tick_params(axis='x', rotation=45)
axes[1].tick_params(axis='x', rotation=45)
axes[2].tick_params(axis='x', rotation=45)

# scale qilish boxplotlarni 
axes[1].set_yscale('log')
axes[2].set_yscale('log')

# umumiy title
fig.suptitle("UFO Sightings: Shape, Duration and Light/Glow Analysis", fontsize=16, fontweight='bold', color='red')

plt.tight_layout()
plt.show()
```

#### Heatmap
